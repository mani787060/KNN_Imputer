```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
#df = pd.read_csv('Titanic-Dataset.csv', usecols=['Age','Pclass','Fare','Survived'])
#or,
df = pd.read_csv('Titanic-Dataset.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean()*100   # for finding % of missing values
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
# call train test split method

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
# now, apply KNN imputer here

knn = KNNImputer(n_neighbors=3, weights='distance')

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
# check for accuracy after that

lr = LogisticRegression()

lr.fit(X_train_trf,y_train)
y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

<h4> what if do here MEAN_IMPUTATION(using simple_imputer class)</h4>

```python
# Comparision it with SimpleImputer --> Mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)  
X_test_trf2 = si.transform(X_test) 
```

```python
# now, finding it's accuracy_score 

lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)
y_pred2 = lr.predict(X_test_trf2)         

accuracy_score(y_test,y_pred2)                                 
```

```python
# That's why, KNN_Imputation is much better technique to impute values comparision to others(mean_median_imputation, etc..)       
```
